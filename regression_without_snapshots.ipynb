```python
import pandas as pd
import statsmodels.api as sm
import math
import numpy as np
```

```python
reds = pd.read_csv("RED_faults.csv")
greys = pd.read_csv("GREY_faults.csv")
```

```python
reds = reds[['odometer_reading', 'message_date_time', 'in_service_dt', 'vehicle_make', 'engine_hours', 'idle_engine_hours', 'gross_combined_weight']]
greys = greys[['odometer_reading', 'message_date_time', 'in_service_dt', 'vehicle_make', 'engine_hours', 'idle_engine_hours', 'gross_combined_weight']]
```

```python
display(reds)
display(greys)
```

```python
reds['red_fault'] = 1
greys['red_fault'] = 0
```

```python
display(reds)
display(greys)
```

```python
faults = pd.concat([reds, greys])
```

```python
display(faults)
```

```python
faults = faults.dropna()
```

```python
display(faults)
```

```python
# Transforms the idle_engine_hours to hours
faults['idle_engine_hours'] = faults['idle_engine_hours'] / 3600
```

```python
# Calculates the active_engine_hours
faults['active_engine_hours'] = faults['engine_hours'] - faults['idle_engine_hours']
print(faults['active_engine_hours'])
```

```python
# Calculate the vehicle service age in days
faults['vehicle_service_age'] = pd.to_datetime(faults['message_date_time']) - pd.to_datetime(faults['in_service_dt'])
faults['vehicle_service_age'] = faults['vehicle_service_age'].dt.days
print(faults['vehicle_service_age'])
```

```python
# Turns vehicle make into a boolean
volvor = lambda x: 1 if x == "VOLVO" else 0
faults['volvo'] = faults['vehicle_make'].apply(volvor)
print(faults['volvo'].mean())
```

```python
display(faults)
```

```python
# The variables for the logistic regression
X = faults[['volvo', 'active_engine_hours', 'gross_combined_weight', 'odometer_reading', 'vehicle_service_age']]
Y = faults['red_fault']
```

```python
display(faults)
```

```python
log_reg = sm.Logit(Y, X).fit()
print(log_reg.summary())
```

```python
coefs = sm.Logit(Y, X).fit().params.values
print(coefs)
```

```python
delta = []
for coef in coefs:
    delta.append((math.exp(coef) - 1) * 100)
print(delta)
```

```python
delta = []
for coef in coefs:
    delta.append((math.exp(coef) - 1) * 100)
print(delta)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
model = LogisticRegression()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=17)
model.fit(X_train, Y_train)
model.predict(X_test)
model.score(X_test, Y_test)
```

```python
confusion_matrix = confusion_matrix(Y_test, model.predict(X_test))
```
