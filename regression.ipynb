```python
import pandas as pd
import statsmodels.api as sm
import math
import numpy as np
```

```python
reds = pd.read_csv("RED_faults.csv")
```

```python
counter = 1
for i in range(len(reds)):
    if reds.vehicle_id[i+1] != reds.vehicle_id[i]:
        counter += 1
    reds["fault_id"][i+1] = counter

display(reds)
```

```python
reds_pvt = reds.pivot_table(columns = 'description', index = 'vehicle_id', values = 'item_value', aggfunc="mean").reset_index()
```
