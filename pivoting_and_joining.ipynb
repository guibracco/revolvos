```python
import pandas as pd
```

```python
# Read in the snaps data
# df = pd.read_csv("RED_snaps.csv")
# df = pd.read_csv("GREY_snaps.csv")
# df = pd.read_csv("ORANGE_snaps.csv")
df = pd.read_csv("BLACK_snaps.csv")
```

```python
# Read in the data without snaps
# df_no_snaps = pd.read_csv("RED_all_faults.csv")
# df_no_snaps = pd.read_csv("GREY_all_faults.csv")
# df_no_snaps = pd.read_csv("ORANGE_all_faults.csv")
df_no_snaps = pd.read_csv("BLACK_all_faults.csv")
```

```python
with pd.option_context('display.max_columns', None):
    display(df)
```

# Pivoting

```python
# Pivot df avoiding duplicates
df_p = df.pivot_table(index="fault_alarm_id", columns="description", values="item_value", aggfunc='first').reset_index()
```

```python
with pd.option_context('display.max_columns', None):
    display(df_p)
```

# Joining

```python
# Join df_p with df
df_all = df_p.join(df_no_snaps.set_index("fault_alarm_id"), on="fault_alarm_id")
```

```python
with pd.option_context('display.max_columns', None):  # more options can be specified also
    display(df_all)
```
