```python
import pandas as pd

df = pd.read_csv("GREY_snaps005.csv")
```

```python
with pd.option_context('display.max_columns', None):
    display(df)
```

# Pivoting

```python
df_p = df.pivot(columns="description", index="fault_alarm_id", values="item_value").reset_index()
```

# Joining

```python
df_all = df_p.join(df.set_index("fault_alarm_id"), on="fault_alarm_id")
```

```python
with pd.option_context('display.max_columns', None):  # more options can be specified also
    display(df_all)
```
